In this notebook I want to learn some things about historical economic factors of Mauritius.

Information on the dataset: 
- The data comes from multiple sources including statistics mauritius, world bank, and investing.com.  
- Inflation has been calculated from average CPI year on year changes with base year 1971.  
- usd is the price of 1 usd in MUR taken for Jan 1st of each year.  
- Unemployment rate was taken from world bank.  
- I was unable to find the MUR exchange rate for the years earlier than 1996, sadly, so they were replaced with Nas.  
- All rates are in percentages. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("datasets/economy.ods")
display(df.describe())
display(df.head(10))
```

I wanted to find what kind of correlation there might be between the different economic indicators. So, I plotted them against each other in different colors. 

Here's a plot of unemployment rate, population growth rate, inflation rate v/s year

```python
years = df["Year"]
unem = df["unemployment_rate"]
popg = df["pop_growth_rate"]
inf = df["Inflation_rate"]
usd = df["usd"]
ys = [unem, popg, inf]
labels = ["unemployment_rate", "pop_growth_rate", "Inflation_rate"]
fig, ax = plt.subplots()
for i in range(len(ys)):
    ax.plot(years, ys[i], label = labels[i])

plt.title("Economic factors")
plt.ylabel("rate %")
plt.xlabel("Year")
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.9))
plt.show()
```
